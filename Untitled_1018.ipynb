```python
#共通関数群

def file_ditect(key_dir):#指定ディレクトリに存在するファイルパスの一覧取得
    path = root + "/" + key_dir + "/" + ym
    
    files = []

    # パスで指定したファイルの一覧をリスト形式で取得
    files = glob.glob(path + '/*.csv')

    #print(files)

    for file in files:
        print(file)
    
    return files
        
def can_load(file):#CANのショップ名割り当て
        can_df = pd.read_csv(file)
        
        can_usual = can_df.query("販売ステータス != 'アウトレット'")
        can_usual = can_usual.replace({"STRIPE CLUB":"CAN"})
        
        can_outlet = can_df.query("販売ステータス == 'アウトレット'")
        can_outlet = can_outlet.replace({"STRIPE CLUB":"CAN OUTLET"})
        
        can = pd.concat([can_usual, can_outlet],axis=0, sort=True)
        
        return can

def stripe_load(file):#ショップ名"KOE"を"STRIPE CLUB"に、ブランド名を"Melan Cleuge"に変換
    st = pd.read_csv(file)
    st = st.replace({"KOE":"STRIPE CLUB"})
    st = st.replace({"Melan Cleuge Men's":"Melan Cleuge", "Melan Cleuge Women's":"Melan Cleuge"})
    
    return st

def palemoba_load(file):#月の調整
    pm = pd.read_csv(file)
    pm_monthPlus = pm.query('日 >= 21')
    pm_monthPlus.loc[:, "月"] += 1
    print(max(pm_monthPlus.loc[:, "月"]))
    pm_over = pm_monthPlus.query('月 == 13')
    pm_over.loc[:, "年"] += 1
    pm_over.loc[:, "月"] -= 12
    print(max(pm_over.loc[:, "月"]))
    pm_monthPlus_noChange = pm_monthPlus.query('月 != 13')
    pm_noChange = pm.query('日 < 21')
    
    pm = pd.concat([pm_over, pm_monthPlus_noChange, pm_noChange], axis = 0)
    
    return pm

def import_csv(file):
    if "CAN" in file:
        df = can_load(file)
    elif "stripe" in file:
        df = stripe_load(file)
    elif "PALEMOBA" in file:
        df = palemoba_load(file)
    else:
        df = pd.read_csv(file)

    df = df[~df["ブランド名"].str.contains("ノベルティ")]
    
    return df
```

```python
#sell用の関数群

'''def import_sell(files):
    #csvファイルの中身を追加していくリストを用意
    data_list = []

    #読み込むファイルのリストを走査
    for file in files:
        #sell用の例外処理
        if "CAN" in file:
            can = can_load(file)
            can.rename(columns={' 出荷金額': '出荷金額'}, inplace=True)
            data_list.append(can)
            continue
        if "stripe" in file:
            st = stripe_load(file)
            st.rename(columns={' 出荷金額': '出荷金額'}, inplace=True)
            data_list.append(st)
            continue

        data_list.append(pd.read_csv(file))

    #リストを全て行方向に結合
    #axis=0:行方向に結合, sort
    df = pd.concat(data_list, axis=0, sort=True)

    #ブランド名にノベルティを含む行の削除
    df = df[~df["ブランド名"].str.contains("ノベルティ")]
    
    #データ型の変換
    df[["出荷金額", "返品金額"]] = df[["出荷金額", "返品金額"]].astype(np.int64)

    #df.columns
    
    return df'''

def shop_sell(df):
    agg = df.query("ショップ名 in @shopUnit")
    
    if agg.empty:
        return
    
    agg = agg.groupby(["ショップ名"]).agg({'出荷金額':'sum','返品金額':'sum','出荷金額-返品金額':'sum'})

    agg.rename(columns={'出荷金額-返品金額': '売上金額'}, inplace=True)
    
    agg = agg.stack()
    
    return agg

def used_sell(df):
    used = df.query('ショップ名 in @usedUnit')
    
    if used.empty:
        return
    
    used = used.query('商品種別 == "USED"')
    
    used_agg = used.groupby(["ショップ名", "ブランド名"]).agg({'出荷金額-返品金額':'sum'})

    used_agg.rename(columns={'出荷金額-返品金額': 'USED売上金額'}, inplace=True)
    
    used_agg = used_agg.stack()

    return used_agg

def brand_sell(df):
    agg_brand = df.query("ショップ名 in @brandUnit")
    
    if agg_brand.empty:
        return

    agg_brand = agg_brand.groupby(["ショップ名", "ブランド名"]).agg({'出荷金額':'sum','返品金額':'sum','出荷金額-返品金額':'sum'})

    agg_brand.rename(columns={'出荷金額-返品金額': '売上金額'}, inplace=True)
    
    agg_brand = agg_brand.stack()

    return agg_brand

def subbrand_sell(df):
    agg_subbrand = df.query("ショップ名 in @subbrandUnit")
    
    if agg_subbrand.empty:
        return

    agg_subbrand = agg_subbrand.groupby(["ショップ名", "ブランド名", "サブブランド名"]).agg({'出荷金額':'sum','返品金額':'sum','出荷金額-返品金額':'sum'})

    agg_subbrand.rename(columns={'出荷金額-返品金額': '売上金額'}, inplace=True)
    
    agg_subbrand = agg_subbrand.stack()

    return agg_subbrand
```

```python
#sell処理
files_sell = file_ditect("sell")
sell_list = []
for file in files_sell:
    sell = import_csv(file)
    df_shop_sell = shop_sell(sell)
    df_brand_sell = brand_sell(sell)
    df_used_sell = used_sell(sell)
    df_subbrand_sell = subbrand_sell(sell)
    sell_all = pd.concat([df_shop_sell, df_brand_sell, df_subbrand_sell, df_used_sell])
    sell_list.append(sell_all)
```

```python
#order用の関数群
def import_order(file):
    df = import_csv(file)    
    #とりあえず変換
    df = df.replace({"セール（シークレット）":"セール", 'アウトレット':"セール"})
    
    return df

def shop_order(df):
    df = df.query("ショップ名 in @shopUnit")
    
    if df.empty:
        return
    
    agg = df.groupby(["ショップ名"]).agg({'受注番号':'nunique','受注数':'sum'})
    agg.rename(columns={'受注番号': '受注件数'}, inplace=True)
    
    agg_sellStatus = df.groupby(["ショップ名", "販売ステータス"]).agg({'受注金額':'sum'})
    
    agg_category_L = df.groupby(["ショップ名", "大カテゴリ名"]).agg({'受注金額':'sum'})
    agg_category_L = agg_category_L.query("ショップ名 in @category_L")
    
    agg_category_M = df.groupby(["ショップ名", "中カテゴリ名"]).agg({'受注金額':'sum'})
    agg_category_M = agg_category_M.query("ショップ名 in @category_M")
    
    agg_orderForm = df.replace({"PC":"PC + MB", 'MB':"PC + MB"})
    agg_orderForm = agg_orderForm.groupby(["ショップ名", "受注形態"]).agg({'受注番号':'nunique'})
    agg_orderForm = agg_orderForm.query("ショップ名 in @orderForm_sum")
    agg_orderForm.rename(columns={'受注番号': '受注件数'}, inplace=True)
    
    agg_purStatus = df.query(("ショップ名 in @purchaseStatus"))
    agg_purStatus = agg_purStatus.groupby(["ショップ名", "購入ステータス"]).agg({'受注番号':'nunique'})
    agg_purStatus.rename(columns={'受注番号': '受注件数'}, inplace=True)
    
    return agg, agg_sellStatus, agg_category_L, agg_category_M, agg_orderForm, agg_purStatus
    
def used_order(df):
    used = df.query('商品種別 == "USED" & ショップ名 in @usedUnit')

    used_agg = used.groupby(["ショップ名", "ブランド名"]).agg({'受注金額':'sum'})

    return used_agg

def brand_order(df):
    df = df.query("ショップ名 in @brandUnit")
    
    agg = df.groupby(["ショップ名", "ブランド名"]).agg({'受注番号':'nunique','受注数':'sum'})
    agg.rename(columns={'受注番号': '受注件数'}, inplace=True)
    
    agg_sellStatus = df.groupby(["ショップ名", "ブランド名", "販売ステータス"]).agg({'受注金額':'sum'})
    
    agg_category_L = df.query("ショップ名 in @category_L")
    agg_category_L = agg_category_L.groupby(["ショップ名", "ブランド名", "大カテゴリ名"]).agg({'受注金額':'sum'})
    
    agg_orderForm_def = df.query("ショップ名 in @orderForm_default")
    agg_orderForm_sum = df.query("ショップ名 in @orderForm_sum")
    agg_orderForm_sum = agg_orderForm_sum.replace({"PC":"PC + MB", 'MB':"PC + MB"})
    agg_orderForm = pd.concat([agg_orderForm_def, agg_orderForm_sum], axis = 0, sort = True)
    agg_orderForm = agg_orderForm.groupby(["ショップ名", "ブランド名", "受注形態"]).agg({'受注番号':'nunique'})
    
    agg_orderForm.rename(columns={'受注番号': '受注件数'}, inplace=True)
    
    agg_purStatus = df.query(("ショップ名 in @purchaseStatus"))
    agg_purStatus = agg_purStatus.groupby(["ショップ名", "ブランド名", "購入ステータス"]).agg({'受注番号':'nunique'})
    agg_purStatus.rename(columns={'受注番号': '受注件数'}, inplace=True)
    
    return agg, agg_sellStatus, agg_category_L, agg_orderForm, agg_purStatus

def subbrand_order(df):
    df = df.query("ショップ名 in @subbrandUnit")
    
    agg = df.groupby(["ショップ名", "ブランド名", "サブブランド名"]).agg({'受注番号':'nunique','受注数':'sum'})
    agg.rename(columns={'受注番号': '受注件数'}, inplace=True)
    
    #agg
    
    agg_sellStatus = df.groupby(["ショップ名", "ブランド名", "サブブランド名", "販売ステータス"]).agg({'受注金額':'sum'})
    
    agg_category_L = df.groupby(["ショップ名", "ブランド名", "サブブランド名", "大カテゴリ名"]).agg({'受注金額':'sum'})
    
    agg_purStatus = agg_purStatus.groupby(["ショップ名", "ブランド名", "サブブランド名", "購入ステータス"]).agg({'受注番号':'nunique'})
    agg_purStatus.rename(columns={'受注番号': '受注件数'}, inplace=True)
    
    return agg, agg_sellStatus, agg_category_L, agg_purStatus
```

```python
#request関数群
def import_request(file):
    df = import_csv(file)
    
    df["リクエスト金額"] = df["定価"]*df["登録数"]
    
    return df

def shop_request(df):
    shopUnit_ex = shopUnit + ["Ailand", "URBAN RESEARCH OUTLET"]
    df = df.query("ショップ名 in @shopUnit_ex")
    df_reg = df.query("登録数 != 0")
    df_send = df.query("送信数 != 0")
    
    agg_reg = df_reg.groupby(["ショップ名"]).agg({'商品コード':'nunique','リクエスト金額':'sum'})
    agg_send = df_send.groupby(["ショップ名"]).agg({'商品コード':'nunique'})
    agg_reg.rename(columns={'商品コード': '登録品番数'}, inplace=True)
    agg_send.rename(columns={'商品コード': '送信品番数'}, inplace=True)
    
    agg = pd.merge(agg_reg, agg_send, on='ショップ名')
    
    return agg

def brand_request(df):
    bU = brandUnit.copy()
    for i in ["Ailand", "URBAN RESEARCH OUTLET", "STRIPE CLUB", "CAN", "CAN OUTLET"]:
        bU.remove(i)
    df = df.query("ショップ名 in @bU")
    df_reg = df.query("登録数 != 0")
    df_send = df.query("送信数 != 0")
    
    agg_reg = df_reg.groupby(["ショップ名", "ブランド名"]).agg({'商品コード':'nunique','リクエスト金額':'sum'})
    agg_send = df_send.groupby(["ショップ名", "ブランド名"]).agg({'商品コード':'nunique'})
    agg_reg.rename(columns={'商品コード': '登録品番数'}, inplace=True)
    agg_send.rename(columns={'商品コード': '送信品番数'}, inplace=True)
    
    agg = pd.merge(agg_reg, agg_send, on='ブランド名')

    return agg

def subbrand_request(df):
    df = df.query("ショップ名 in @subbrandUnit")
    
    agg = df.groupby(["ショップ名", "ブランド名", "サブブランド名"]).agg({'品番数':'sum','日':'nunique'})
    agg.rename(columns={'日': 'アップ日数'}, inplace=True)

    return agg
```
