```python
import os
import datetime
import pandas as pd
import numpy as np
import glob

#表示行数の増加
pd.set_option('display.max_rows', 500)

today = datetime.datetime.now()

y = today.year
m = today.month - 1

ym = str(y) + str('{0:02d}'.format(m))

#暫定でのroot dir
root = "/Users/miyamoto/Documents/Root/raw/daily_sell"
```

```python
def file_ditect(key_dir):#指定ディレクトリに存在するファイルパスの一覧取得
    path = root + "/" + key_dir + "/"
    
    files = []

    # パスで指定したファイルの一覧をリスト形式で取得
    files = glob.glob(path + '/*.csv')

    #print(files)

    for file in files:
        print(file)
    
    return files
```

```python
def import_sell(file):
    dataframe = pd.read_csv(file)
    '''dataframe["年"] = pd.to_datetime(dataframe["日付"]).dt.year.astype(str)
    dataframe["月"] = pd.to_datetime(dataframe["日付"]).dt.month.astype(str)
    dataframe["日"] = pd.to_datetime(dataframe["日付"]).dt.day.astype(str)'''
    
    dataframe['年月'] = dataframe["年"].astype(str)+"年"+dataframe["月"].astype(str)+"月"
    
    return dataframe

def aggregate(df):
    df_agg = df.groupby(["ショップ名", "年月"]).agg({"出荷金額-返品金額":"sum"})
    return df_agg
```

```python
files_sell = file_ditect("")
```

```python
sell_list_shop = []

for file in files_sell:
    print(file)
    sell = import_sell(file)
    df_shop_sell = aggregate(sell)
    sell_list_shop.append(df_shop_sell)

sh_sell = pd.concat(sell_list_shop,axis = 0).reset_index()
sh_sell = sh_sell[sh_sell.columns[sh_sell.columns != 'index']]
sh_sell.rename(columns={"出荷金額-返品金額":"売上"}, inplace =True) 
```

```python
sh_sell_piv = pd.pivot_table(sh_sell, index="ショップ名", columns="年月", values="売上")
```
